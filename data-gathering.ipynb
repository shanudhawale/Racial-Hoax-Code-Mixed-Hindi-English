```python
import os
import json
import pandas as pd

json_dir = '/content/drive/MyDrive/Thesis2'

all_comments = []

for file_name in os.listdir(json_dir):
    if file_name.endswith('.json'):
        file_path = os.path.join(json_dir, file_name)
        print(file_path)
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
            for comment in data.get('comments', []):
              cid = comment.get('cid', '')
              author = comment.get('author', '')
              text = comment.get('text', '')
              replies = comment.get('replies', '')
              reply = comment.get('reply','')
              all_comments.append([cid, author, text, replies,reply])
```

```python
import csv

with open('output.csv', 'w', newline='', encoding='utf-8') as csv_file:
  writer = csv.writer(csv_file)
  # Write the header
  writer.writerow(['cid','author', 'text', 'replies','reply'])
  # Write the data
  writer.writerows(all_comments)
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("token-classification", model="l3cube-pune/hing-bert-lid", device=0)

def detect_eng_in_hindi(text):
    result = pipe(text)
    is_english = all(list(j.values())[0] == 'EN' for j in result)
    return 1 if is_english else 0


```
